```python
import pandas as pd
import os
from collections import defaultdict
from scipy.io import arff
```

```python
# check if a folder (directory) exists

fill_dirc = '/Users/chengtao/Documents/GitHub/CS470-Team-10/phase1/NATOPS'

if os.path.exists(fill_dirc):
    print(f"The folder at '{fill_dirc}' exists.")
else:
    print(f"The folder at '{fill_dirc}' does not exist.")
```

```python
file_list_test = [f for f in os.listdir(fill_dirc) if (f.startswith('NATOPSDimension') and 'TEST' in f)]
```

```python
file_list_test = sorted(file_list_test)
```

```python
file_list_train = [f for f in os.listdir(fill_dirc) if (f.startswith('NATOPSDimension') and 'TRAIN' in f)]
```

```python
file_list_train = sorted(file_list_train)
```

# training dataset

```python
datasets_train =[]
```

```python
for file_name in file_list_train:
    file_path = os.path.join(fill_dirc, file_name)
    steps, meta = arff.loadarff(file_path)

    epoch = pd.DataFrame(steps)

    datasets_train.append(epoch)
    
for epoch in datasets_train:
    epoch.insert(0, 'sid', range(1, 181))
    epoch.insert(1, 'data_type', 0)
```

```python
datasets_train[0]
```

```python
expanded_columns_train = []
column_name = []


for epoch in datasets_train:
    selected_data = epoch.iloc[:, 2:53]
    
    column_name.append(epoch.columns[2].split('_')[1])

    expanded_data = []
    
    for _, row in selected_data.iterrows():
        
        expanded_data.extend(row.tolist())
        
    
    expanded_columns_train.append(expanded_data)
        
final_dataframe_train = pd.DataFrame({f'Dimension_{i + 1}':col for i, col in enumerate(expanded_columns_train)})
```

```python
column_mapping = {f'Dimension_{i}': f'Dimension_{column_name[i]}' for i in range(24)}

final_dataframe_train.rename(columns=column_mapping, inplace=True)
final_dataframe_train
```

```python
index_values = []
index_ranges = [(i // 51) + 1 for i in range(9180)]

for index in index_ranges:
    index_values.extend([index])
    
final_dataframe_train.insert(0, 'sid', index_values)

final_dataframe_train.insert(1, 'data_type', 0)
```

```python
class_values = []

for _, row in final_dataframe_train.iterrows():
    
    sample_id = row['sid']
    
    class_value = datasets_train[0].loc[datasets_train[0]['sid'] == sample_id, 'classAttribute'].values[0]
    
    class_values.append(class_value)
    
final_dataframe_train['classAttribute'] = class_values

final_dataframe_train
```

# testing dataset

```python
datasets_test = []
```

```python
# Read and convert each ARFF file to DataFrame
for file_name in file_list_test:
    file_path = os.path.join(fill_dirc, file_name)
    steps, meta = arff.loadarff(file_path)

    epoch = pd.DataFrame(steps)

    datasets_test.append(epoch)
    
for epoch in datasets_test:
  
    epoch.insert(0, 'sid', range(1, 181))
    epoch.insert(1, 'data_type', 1)
```

```python
datasets_test
```

```python
expanded_columns_test = []
column_name_test = []

for epoch in datasets_test:
    selected_data = epoch.iloc[:, 2:53]
    
    column_name_test.append(epoch.columns[2].split('_')[1])

    expanded_data = []
    
    for _, row in selected_data.iterrows():
        
        expanded_data.extend(row.tolist())
        
    
    expanded_columns_test.append(expanded_data)
    
final_dataframe_test = pd.DataFrame({f'Dimension_{i + 1}':col for i, col in enumerate(expanded_columns_test)})
```

```python
column_mapping_test = {f'Dimension_{i}': f'Dimension_{column_name[i]}' for i in range(24)}

final_dataframe_test.rename(columns=column_mapping_test, inplace=True)
final_dataframe_test
```

```python
index_values_test = []
index_ranges_test = [(i // 51) + 1 for i in range(9180)]

for index in index_ranges_test:
    index_values_test.extend([index])
    
final_dataframe_test.insert(0, 'sid', index_values_test)

final_dataframe_test.insert(1, 'data_type', 1)
```

```python
class_values_test = []

for _, row in final_dataframe_test.iterrows():
    
    sample_id = row['sid']
    
    class_value_test = datasets_test[0].loc[datasets_test[0]['sid'] == sample_id, 'classAttribute'].values[0]
    
    class_values_test.append(class_value_test)
    
final_dataframe_test['classAttribute'] = class_values_test

final_dataframe_test
```

# combine train dataset and test dataset 

```python
final_combined_dataframe = pd.concat([final_dataframe_test, final_dataframe_train], axis=0)
final_combined_dataframe
```
