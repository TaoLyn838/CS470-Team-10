```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
file_path = 'NATOPS_sid20-C_TRAIN_su.csv'  
natops_data = pd.read_csv(file_path)
```

```python
natops_data
```

```python
data_processed = natops_data.drop(['isTest'], axis=1)
```

```python
data_processed = data_processed.drop(['sid'], axis=1)
```

```python
data_processed.to_csv("data_processed.csv", index=False)
```

```python
features = data_processed.drop(['class'], axis=1)
```

```python
target = data_processed['class']
```

```python
k_folds = KFold(n_splits = 5)
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier(random_state=42))
])
```

```python
scores = cross_val_score(pipeline, features, target, cv= k_folds)
```
